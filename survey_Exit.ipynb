```python
from scripts import process
from scripts import analyse
```

```python
df, _, _ = process.DataCleaning.clean('survey', 'data', 'output', clear=True, export_df=False)
hos, _, _ = process.DataCleaning.clean('clinic_forQR', 'data', 'output', clear=True, export_df=False)

for var in ['durHome', 'durWaiting', 'durServ']:
    process.DataManipulation.mapping(df, var)

for var in ['satisfyOverall', 'satisfyDistance', 'satisfyProcess', 'satisfyWaiting', 'satisfyProfessional', 'satisfyTrust', 'satisfyEnv', 'satisfyPrivacy']:
    process.DataManipulation.mapping(df, var)
```
